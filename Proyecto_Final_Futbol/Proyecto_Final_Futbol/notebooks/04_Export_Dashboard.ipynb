# 04 — Exportar para Dashboard

```python
import pandas as pd
df=pd.read_csv("data/processed/jugadores_final.csv", sep=";", decimal=",", parse_dates=["fecha_partido"])

agg_jugador=df.groupby(["jugador_id","nombre","equipo","posicion","nacionalidad","temporada"],as_index=False).agg(partidos=("minutos","count"),minutos=("minutos","sum"),goles=("goles","sum"),asistencias=("asistencias","sum"),contrib=("contrib","sum"),xg=("xg","sum"),xa=("xa","sum"),salario_eur=("salario_eur","median"))
for c in ["goles","asistencias","contrib","xg","xa"]:
    agg_jugador[f"{c}_por90"]=agg_jugador[c]/agg_jugador["minutos"]*90
agg_jugador.to_csv("data/processed/exports/jugadores_temporada.csv",index=False,sep=";",decimal=",")

agg_equipo=df.groupby(["equipo","temporada"],as_index=False).agg(goles=("goles","sum"),asistencias=("asistencias","sum"),contrib=("contrib","sum"))
agg_equipo.to_csv("data/processed/exports/equipos_temporada.csv",index=False,sep=";",decimal=",")

agg_pos=df.groupby(["posicion","temporada"],as_index=False).agg(goles=("goles","sum"),asistencias=("asistencias","sum"),contrib=("contrib","sum"))
agg_pos.to_csv("data/processed/exports/posiciones_temporada.csv",index=False,sep=";",decimal=",")

df.sample(min(60000,len(df)),random_state=1).to_csv("data/processed/exports/muestra_detalle_partidos.csv",index=False,sep=";",decimal=",")
print("Exports listos en data/processed/exports/")
```
