# 01 — Generar Dataset Fútbol Sintético

```python
import pandas as pd
contratos = pd.read_csv("data/raw/contratos_jugadores_raw.csv", sep=";", decimal=",")
estadisticas = pd.read_csv("data/raw/estadisticas_partidos_raw.csv", sep=";", decimal=",", parse_dates=["fecha_partido"])
df = estadisticas.merge(contratos, on="jugador_id", how="left")

# Features
df["imc"]=df["peso_kg"]/(df["altura_cm"]/100)**2
df["contrib"]=df["goles"]+df["asistencias"]
for c in ["contrib","goles","asistencias","tiros","pases_completados","xg","xa"]:
    df[f"{c}_por90"]=df[c]/df["minutos"]*90
df["salario_millones"]=df["salario_eur"]/1_000_000

df.to_csv("data/processed/jugadores_final.csv", index=False, sep=";", decimal=",")
df.head()
```
