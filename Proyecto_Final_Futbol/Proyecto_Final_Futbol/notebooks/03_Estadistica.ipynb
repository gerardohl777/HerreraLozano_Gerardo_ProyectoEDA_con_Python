# 03 — Estadística

```python
import pandas as pd
from scipy import stats as st
df=pd.read_csv("data/processed/jugadores_final.csv", sep=";", decimal=",")

# ANOVA contrib_por90 por posición
groups=[g["contrib_por90"].dropna() for _,g in df.groupby("posicion")]
f,p=st.f_oneway(*groups)
print("ANOVA contrib_por90 ~ posicion: F=%.3f p=%.4f"%(f,p))

# Correlación edad vs contrib_por90
d=df[["edad","contrib_por90"]].dropna()
r,p=st.pearsonr(d["edad"],d["contrib_por90"])
print("Correlación Pearson edad vs contrib_por90: r=%.3f p=%.4f"%(r,p))

# t-test pie izquierdo vs derecho
izq=df.loc[df["pie_preferido"]=="Left","contrib_por90"].dropna()
der=df.loc[df["pie_preferido"]=="Right","contrib_por90"].dropna()
t,p=st.ttest_ind(izq,der,equal_var=False)
print("t-test Left vs Right contrib_por90: t=%.3f p=%.4f"%(t,p))
```
