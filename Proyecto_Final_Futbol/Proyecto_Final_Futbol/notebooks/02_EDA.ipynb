# 02 — EDA Jugadores Fútbol

```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, os
os.makedirs("reports/figures", exist_ok=True)
df = pd.read_csv("data/processed/jugadores_final.csv", sep=";", decimal=",", parse_dates=["fecha_partido"])
print("Shape:", df.shape)
```

## Histogramas

```python
for c in ["minutos","goles","asistencias","contrib_por90","salario_eur","imc"]:
    plt.figure()
    df[c].dropna().hist(bins=50)
    plt.title(c)
    plt.savefig(f"reports/figures/hist_{c}.png")
    plt.show()
```

## Boxplots por posición

```python
for c in ["contrib_por90","goles_por90","asistencias_por90"]:
    plt.figure()
    df.boxplot(column=c,by="posicion")
    plt.title(c)
    plt.suptitle("")
    plt.savefig(f"reports/figures/box_{c}_by_pos.png")
    plt.show()
```

## Correlación

```python
num=df.select_dtypes("number").corr()
plt.figure(figsize=(10,8))
sns.heatmap(num,cmap="coolwarm")
plt.savefig("reports/figures/corr_matrix.png")
plt.show()
```
